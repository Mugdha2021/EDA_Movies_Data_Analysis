**6.1 Handling Missing data**

```python
# Insert your code here...
null_frame = pd.DataFrame(index = data.columns.values)
null_frame['Null Frequency']= data.isnull().sum().values
percent = data.isnull().sum().values/data.shape[0]
null_frame['Missing data%']= np.round(percent,decimals = 4) * 100
null_frame.transpose()
```

```python
data.groupby(['Year'])['Revenue (Millions)'] .agg(['count','sum','mean','median','std'])
```

```python
data[data['Revenue (Millions)'].isnull()]
```

**Observations Made**

- We can observe following features are found to have missing values along with its proportion

|Feature|Object Type|Missing Values|Missing Proportion|Solution|
|:--:|:--:|:--:|:--:|:--|
|Revenue (Millions)|float64|128|12.8%|Replace with median.|
|Metascore|float64|64|6.4%|Replace with mean.|


- Here, since the revenue can be affected by year, we need to follow pandas groupby method
- **Replacing null values in columns Revenue (Millions)** with **median of year** 

```python
data_revenue_median = data.groupby('Year')['Revenue (Millions)'].median()
data['Revenue (Millions)'].fillna(data['Year'].map(data_revenue_median), inplace=True)
data
```

```python
data[data['Revenue (Millions)'].isnull()]
```

- Metascore is the rating provided by best critics.
- Metascore has 64 missing values.
- We will replace the Metascore with the **Mean of Year**

```python
data_metascore_mean = data.groupby('Year')['Metascore'].mean()
data['Metascore'].fillna(data['Year'].map(data_metascore_mean), inplace=True)
data
```

```python
data[data['Metascore'].isnull()]
```

- Lets verify the integrity of null values again

```python
null_frame = pd.DataFrame(index = data.columns.values)
null_frame['Null Frequency']= data.isnull().sum().values
percent = data.isnull().sum().values/data.shape[0]
null_frame['Missing data%']= np.round(percent,decimals = 4) * 100
null_frame.transpose()
```

- The objective of handling missing data is accomplished
